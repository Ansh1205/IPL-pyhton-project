Visualizations of IPL Data 

```python
winners = df['winner'].value_counts()
ax = winners.plot(kind='bar', color='skyblue', figsize=(10, 6))
for i, count in enumerate(winners):
    ax.text(i, count + 0.05, str(count), ha='center', va='bottom')
plt.title("Winners Counts")
plt.xlabel("Teams")
plt.ylabel("Counts")
plt.show()
```

```python
MOTM = df['player_of_match'].value_counts().head(10)
ax = MOTM.plot(kind = 'barh' ,color='red', figsize=(10, 6))
for i, count in enumerate(MOTM):
    ax.text(count + 0.5, i, str(count), va='center')
plt.show()
```

```python
#venues = df['venue'].nunique()
venue_data = df[df['venue'] == 'M Chinnaswamy Stadium' ]
toss_outcome_counts = venue_data['toss_decision'].value_counts()
max_outcome = toss_outcome_counts.idxmax()
max_outcome_count = toss_outcome_counts.max()
print(f"The maximum outcome of the toss at {'M Chinnaswamy Stadium'}is {max_outcome} with {max_outcome_count} occurrences.")
```

```python
field = (df['toss_decision'] == 'field').sum()
bat = (df['toss_decision'] == 'bat').sum()
sizes = [field, bat]
labels = ['Field', 'Bat']
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.show()
```
